## Sistema de Chamada Interativo

**‚ö†Ô∏è IMPORTANTE:** Antes de executar o c√≥digo abaixo, certifique-se de que a pasta `data` existe.
Execute a c√©lula abaixo para criar a pasta automaticamente.

```python
# Criar pasta data automaticamente
# Usando magic command do Jupyter (!mkdir)
# O comando ! executa comandos do sistema diretamente no notebook

import os
os.makedirs('data', exist_ok=True)

print("‚úì Pasta 'data' criada/verificada com sucesso!")
```

```python
# Fun√ß√µes para gerenciar disciplinas usando apenas arquivos texto simples
# (sem usar m√≥dulos json, os, datetime, threading, time)

def carregar_disciplinas():
    """Carrega disciplinas do arquivo texto."""
    disciplinas = {}
    linhas = []
    
    # Cria o arquivo vazio se n√£o existir usando modo 'a' (append)
    # O modo 'a' cria o arquivo automaticamente se ele n√£o existir
    with open('data/disciplinas.txt', 'a', encoding='utf-8') as f:
        pass  # Apenas cria o arquivo se n√£o existir
    
    # Agora l√™ o arquivo (que sempre existe agora)
    with open('data/disciplinas.txt', 'r', encoding='utf-8') as f:
        linhas = f.readlines()
    
    # Processa cada linha do arquivo
    i = 0
    while i < len(linhas):
        # Cada disciplina tem 3 linhas: nome, num_encontros, e linha separadora
        if i < len(linhas):
            nome = linhas[i].strip()
            if nome and nome != "---":  # Se n√£o for linha vazia ou separadora
                if i + 1 < len(linhas):
                    num_encontros_str = linhas[i + 1].strip()
                    # Verifica se √© um n√∫mero v√°lido usando m√©todo isdigit()
                    if num_encontros_str.isdigit():
                        num_encontros = int(num_encontros_str)
                        disciplinas[nome] = {
                            'nome': nome,
                            'num_encontros': num_encontros,
                            'chamadas': {}
                        }
                        i += 3  # Pula nome, num_encontros e linha separadora
                    else:
                        i += 1
                else:
                    break
            else:
                i += 1
        else:
            break
    
    return disciplinas

def salvar_disciplinas(disciplinas):
    """Salva disciplinas no arquivo texto."""
    # Abre o arquivo em modo escrita
    with open('data/disciplinas.txt', 'w', encoding='utf-8') as f:
        # Escreve cada disciplina no formato: nome\nnum_encontros\n---\n
        for nome, info in disciplinas.items():
            f.write(f"{info['nome']}\n")
            f.write(f"{info['num_encontros']}\n")
            f.write("---\n")  # Linha separadora

def listar_disciplinas():
    """Lista todas as disciplinas cadastradas."""
    disciplinas = carregar_disciplinas()
    return disciplinas

def criar_disciplina(nome, num_encontros):
    """Cria uma nova disciplina."""
    disciplinas = carregar_disciplinas()
    disciplinas[nome] = {
        'nome': nome,
        'num_encontros': num_encontros,
        'chamadas': {}
    }
    salvar_disciplinas(disciplinas)
    return disciplinas[nome]

def salvar_chamada(disciplina_nome, encontro, presencas):
    """Salva o resultado da chamada em arquivo texto."""
    # Salva tamb√©m em arquivo txt separado
    arquivo_txt = f'data/chamada_{disciplina_nome.replace(" ", "_")}_encontro_{encontro}.txt'
    with open(arquivo_txt, 'w', encoding='utf-8') as f:
        # Escreve cabe√ßalho
        f.write("="*70 + "\n")
        f.write(f"CHAMADA - {disciplina_nome.upper()}\n")
        f.write(f"Encontro: {encontro}\n")
        f.write("="*70 + "\n\n")
        
        # Filtra apenas os alunos presentes (status 'P')
        presentes = [p for p in presencas if p['status'] == 'P']
        # Filtra apenas os alunos ausentes (status 'F')
        ausentes = [p for p in presencas if p['status'] == 'F']
        
        # Escreve estat√≠sticas
        f.write(f"TOTAL DE ALUNOS: {len(presencas)}\n")
        f.write(f"PRESENTES: {len(presentes)}\n")
        f.write(f"AUSENTES: {len(ausentes)}\n")
        f.write("\n" + "-"*70 + "\n")
        f.write("PRESENTES:\n")
        f.write("-"*70 + "\n")
        # Itera sobre os alunos presentes, enumerando a partir de 1
        for i, aluno in enumerate(presentes, 1):
            f.write(f"{i:2d}. [{aluno['matricula']}] {aluno['nome']}\n")
        
        f.write("\n" + "-"*70 + "\n")
        f.write("AUSENTES:\n")
        f.write("-"*70 + "\n")
        # Itera sobre os alunos ausentes, enumerando a partir de 1
        for i, aluno in enumerate(ausentes, 1):
            f.write(f"{i:2d}. [{aluno['matricula']}] {aluno['nome']}\n")
    
    print(f"\n‚úÖ Chamada salva em: {arquivo_txt}")
    return arquivo_txt
```

```python
# Sistema de chamada simplificado (sem threading e timeout)
# Usa apenas input() b√°sico

def fazer_chamada(disciplina_nome, encontro, alunos_dict):
    """Realiza a chamada interativa."""
    # Exibe cabe√ßalho da chamada com linha separadora de 70 caracteres
    print("\n" + "="*70)
    # Exibe o nome da disciplina em mai√∫sculas
    print(f"INICIANDO CHAMADA - {disciplina_nome.upper()}")
    # Exibe o n√∫mero do encontro
    print(f"ENCONTRO: {encontro}")
    # Exibe outra linha separadora
    print("="*70)
    # Exibe instru√ß√µes sobre como marcar presen√ßa
    print("\nDigite 'P' ou 'p' para PRESENTE")
    print("Digite qualquer outra coisa ou Enter para AUSENTE")
    # Exibe linha separadora e pula uma linha
    print("\n" + "-"*70 + "\n")
    
    # Inicializa lista para armazenar os resultados da chamada
    presencas = []
    # Converte o dicion√°rio de alunos em lista de tuplas (matricula, nome)
    alunos_lista = list(alunos_dict.items())
    
    # Itera sobre cada aluno, enumerando a partir de 1
    for i, (matricula, nome) in enumerate(alunos_lista, 1):
        # Exibe linha separadora antes de cada aluno
        print("-"*70)
        # Exibe o n√∫mero do aluno atual e total (ex: [1/28])
        print(f"[{i}/{len(alunos_lista)}]")
        # Exibe o nome do aluno de forma destacada
        print(f"ALUNO: {nome}")
        # Exibe a matr√≠cula do aluno
        print(f"MATR√çCULA: {matricula}")
        print("-"*70)
        
        # Solicita resposta do usu√°rio usando input() b√°sico
        # O prompt inclui o nome do aluno para garantir que seja vis√≠vel
        resposta = input(f"O aluno {nome} est√° presente? (P/p = Presente, Enter = Ausente): ")
        
        # Verifica se a resposta √© 'P' (mai√∫sculo ou min√∫sculo)
        if resposta.upper() == 'P':
            # Define status como presente
            status = 'P'
            # Exibe confirma√ß√£o de presen√ßa
            print("‚úÖ PRESENTE registrado")
        else:
            # Define status como faltante
            status = 'F'
            # Exibe mensagem de aus√™ncia
            print("‚ùå AUSENTE registrado")
        
        # Adiciona os dados do aluno √† lista de presen√ßas
        presencas.append({
            'matricula': matricula,  # Matr√≠cula do aluno
            'nome': nome,  # Nome completo do aluno
            'status': status  # Status: 'P' para presente, 'F' para faltante
        })
        
        # Adiciona uma linha em branco entre alunos para melhor visualiza√ß√£o
        print()
    
    # Exibe linha separadora final
    print("\n" + "="*70)
    # Exibe mensagem de conclus√£o
    print("CHAMADA CONCLU√çDA!")
    # Exibe outra linha separadora
    print("="*70)
    
    # Resumo
    # Conta quantos alunos est√£o presentes (status 'P')
    presentes_count = sum(1 for p in presencas if p['status'] == 'P')
    # Calcula quantos alunos est√£o ausentes (total - presentes)
    ausentes_count = len(presencas) - presentes_count
    
    # Exibe cabe√ßalho do resumo
    print(f"\nüìä RESUMO:")
    # Exibe o total de alunos
    print(f"Total de alunos: {len(presencas)}")
    # Exibe o n√∫mero de presentes
    print(f"Presentes: {presentes_count}")
    # Exibe o n√∫mero de ausentes
    print(f"Ausentes: {ausentes_count}")
    # Calcula e exibe a taxa de presen√ßa em porcentagem
    print(f"Taxa de presen√ßa: {(presentes_count/len(presencas)*100):.1f}%")
    
    # Retorna a lista de presen√ßas
    return presencas
```

```python
def menu_principal():
    """Menu principal do sistema de chamada."""
    # Carrega todas as disciplinas cadastradas
    disciplinas = listar_disciplinas()
    
    # Exibe cabe√ßalho do sistema com linha separadora de 70 caracteres
    print("\n" + "="*70)
    # Exibe t√≠tulo do sistema
    print("SISTEMA DE CHAMADA")
    # Exibe outra linha separadora
    print("="*70)
    
    # Listar disciplinas existentes
    # Verifica se existem disciplinas cadastradas
    if disciplinas:
        # Exibe cabe√ßalho da se√ß√£o de disciplinas
        print("\nüìö DISCIPLINAS CADASTRADAS:")
        # Exibe linha separadora com h√≠fens
        print("-"*70)
        # Converte o dicion√°rio de disciplinas em lista de tuplas (nome, info)
        disciplinas_lista = list(disciplinas.items())
        # Itera sobre as disciplinas, enumerando a partir de 1
        for i, (nome, info) in enumerate(disciplinas_lista, 1):
            # Exibe cada disciplina com seu n√∫mero e quantidade de encontros
            print(f"{i}. {nome} ({info['num_encontros']} encontros)")
        # Exibe op√ß√£o para criar nova disciplina
        print(f"{len(disciplinas_lista) + 1}. Nova Disciplina")
        # Exibe linha separadora final
        print("-"*70)
        
        # Exibe resumo das op√ß√µes antes de pedir a escolha
        print("\nOP√á√ïES DISPON√çVEIS:")
        # Lista novamente as op√ß√µes de forma destacada
        for i, (nome, info) in enumerate(disciplinas_lista, 1):
            print(f"  Op√ß√£o {i}: Selecionar disciplina '{nome}'")
        print(f"  Op√ß√£o {len(disciplinas_lista) + 1}: Criar nova disciplina")
        print("-"*70)
        
        # Solicita escolha do usu√°rio
        escolha = input(f"\nEscolha uma op√ß√£o (1 a {len(disciplinas_lista) + 1}): ").strip()
        
        # Verifica se a escolha √© um n√∫mero v√°lido
        escolha_valida = True
        escolha_num = 0
        
        # Tenta converter para n√∫mero inteiro verificando cada caractere
        if escolha.isdigit():
            escolha_num = int(escolha)
        else:
            escolha_valida = False
        
        # Verifica se a escolha est√° dentro do range v√°lido
        if escolha_valida and 1 <= escolha_num <= len(disciplinas_lista):
            # Disciplina existente
            # Obt√©m o nome da disciplina selecionada (√≠ndice - 1 porque lista come√ßa em 0)
            disciplina_nome = disciplinas_lista[escolha_num - 1][0]
            # Obt√©m as informa√ß√µes da disciplina selecionada
            disciplina_info = disciplinas[disciplina_nome]
            # Exibe confirma√ß√£o da disciplina selecionada
            print(f"\n‚úì Disciplina selecionada: {disciplina_nome}")
            # Exibe o total de encontros da disciplina
            print(f"‚úì Total de encontros: {disciplina_info['num_encontros']}")
            
            # Perguntar qual encontro
            # Solicita o n√∫mero do encontro para a chamada
            encontro = input(f"\nPara qual encontro ser√° a chamada? (1 a {disciplina_info['num_encontros']}): ").strip()
            
            # Verifica se √© um n√∫mero v√°lido
            encontro_valido = True
            encontro_num = 0
            
            if encontro.isdigit():
                encontro_num = int(encontro)
            else:
                encontro_valido = False
            
            # Verifica se o n√∫mero do encontro √© v√°lido
            if encontro_valido and 1 <= encontro_num <= disciplina_info['num_encontros']:
                # Fazer chamada
                # Realiza a chamada para o encontro especificado
                presencas = fazer_chamada(disciplina_nome, encontro_num, alunos)
                # Salva os resultados da chamada
                salvar_chamada(disciplina_nome, encontro_num, presencas)
            else:
                # Exibe mensagem de erro se o n√∫mero do encontro for inv√°lido
                print("‚ùå N√∫mero de encontro inv√°lido!")
                # Retorna para encerrar a fun√ß√£o
                return
                    
        elif escolha_valida and escolha_num == len(disciplinas_lista) + 1:
            # Nova disciplina
            # Chama fun√ß√£o para criar uma nova disciplina
            criar_nova_disciplina()
        else:
            # Exibe mensagem de erro se a op√ß√£o escolhida for inv√°lida
            print("‚ùå Op√ß√£o inv√°lida!")
            # Retorna para encerrar a fun√ß√£o
            return
    else:
        # Nenhuma disciplina cadastrada
        # Exibe mensagem informando que n√£o h√° disciplinas cadastradas
        print("\nüìù Nenhuma disciplina cadastrada ainda.")
        # Chama fun√ß√£o para criar uma nova disciplina
        criar_nova_disciplina()

def criar_nova_disciplina():
    """Cria uma nova disciplina e inicia a chamada."""
    # Exibe cabe√ßalho da se√ß√£o de nova disciplina
    print("\n" + "-"*70)
    # Exibe t√≠tulo da se√ß√£o
    print("NOVA DISCIPLINA")
    # Exibe linha separadora
    print("-"*70)
    
    # Solicita o nome da disciplina e remove espa√ßos em branco no in√≠cio e fim
    nome = input("Digite o nome da disciplina: ").strip()
    # Verifica se o nome n√£o est√° vazio
    if not nome:
        # Exibe mensagem de erro se o nome estiver vazio
        print("‚ùå Nome da disciplina n√£o pode estar vazio!")
        # Retorna para encerrar a fun√ß√£o
        return
    
    # Solicita a quantidade de encontros
    num_encontros_str = input("Digite a quantidade de encontros da disciplina: ").strip()
    
    # Verifica se √© um n√∫mero v√°lido
    num_encontros_valido = True
    num_encontros = 0
    
    if num_encontros_str.isdigit():
        num_encontros = int(num_encontros_str)
    else:
        num_encontros_valido = False
    
    # Verifica se o n√∫mero de encontros √© v√°lido
    if not num_encontros_valido:
        # Exibe mensagem de erro se a entrada n√£o for um n√∫mero v√°lido
        print("‚ùå Por favor, digite um n√∫mero v√°lido para a quantidade de encontros!")
        # Retorna para encerrar a fun√ß√£o
        return
    
    # Verifica se o n√∫mero de encontros √© maior que zero
    if num_encontros <= 0:
        # Exibe mensagem de erro se o n√∫mero for inv√°lido
        print("‚ùå O n√∫mero de encontros deve ser maior que zero!")
        # Retorna para encerrar a fun√ß√£o
        return
    
    # Criar disciplina
    # Cria a disciplina com o nome e n√∫mero de encontros informados
    criar_disciplina(nome, num_encontros)
    # Exibe mensagem de sucesso
    print(f"\n‚úÖ Disciplina '{nome}' criada com sucesso!")
    
    # Perguntar qual encontro
    # Solicita o n√∫mero do encontro para a chamada
    encontro = input(f"\nPara qual encontro ser√° a chamada? (1 a {num_encontros}): ").strip()
    
    # Verifica se √© um n√∫mero v√°lido
    encontro_valido = True
    encontro_num = 0
    
    if encontro.isdigit():
        encontro_num = int(encontro)
    else:
        encontro_valido = False
    
    # Verifica se o n√∫mero do encontro √© v√°lido
    if encontro_valido and 1 <= encontro_num <= num_encontros:
        # Fazer chamada
        # Realiza a chamada para o encontro especificado
        presencas = fazer_chamada(nome, encontro_num, alunos)
        # Salva os resultados da chamada
        salvar_chamada(nome, encontro_num, presencas)
    else:
        # Exibe mensagem de erro se o n√∫mero do encontro for inv√°lido
        print("‚ùå N√∫mero de encontro inv√°lido!")
        # Retorna para encerrar a fun√ß√£o
        return

# Executar o sistema
# Verifica se o script est√° sendo executado diretamente (n√£o importado)
if __name__ == "__main__":
    # Chama a fun√ß√£o principal do sistema
    menu_principal()
```

### Como usar o sistema:

2. Escolha uma disciplina existente ou crie uma nova
3. Informe o n√∫mero do encontro
4. Para cada aluno, digite `P` ou `p` para marcar presen√ßa
5. Os resultados s√£o salvos automaticamente em arquivos na pasta `data/`

**Arquivos gerados:**
- `data/disciplinas.txt` - Armazena todas as disciplinas cadastradas (formato texto simples)
- `data/chamada_[DISCIPLINA]_encontro_[N√öMERO].txt` - Relat√≥rio de cada chamada
